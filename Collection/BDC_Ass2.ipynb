Những dạng biểu diễn dữ liệu khuyết trong dataframe

- Hiển thị "NA" hoặc "NaN"

- Ô dữ liệu bị trống

--> Standard missing values (dữ liệu khuyết tiêu chuẩn)

- Hiển thị "na", "n/a

#### 4.1.2. Kiểm tra và xử lý ngoại lệ (outlier)

⚓ Vì sao cần xử lý ngoại lệ?
- Đối với dữ liệu dạng số, các giá trị ngoại lệ có thể là những dữ liệu phi thực tế như số tuổi âm, hoặc những giá trị khác xa so với phần còn lại trong trường giá trị đó, etc.
- Đối với dữ liệu dạng hạng mục, dữ liệu ngoại lệ có thể là một hạng mục phi thực tế, ví dụ như một địa danh không có trên bản đồ, etc.

--> Trong vài trường hợp, các dữ liệu ngoại lệ có thể chứa thông tin có giá trị, nhưng phần đa các dữ liệu ngoại lệ đều vô nghĩa và gây sai lệch trong quá trình xây dựng mô hình

⚓ Một số phương pháp xác định và xử lý giá trị ngoại lệ

1) Xác định giá trị ngoại lệ


> Đối với dữ liệu số:

a) Các giá trị không nằm trong miền xác định của dữ liệu. Ví dụ: tuổi không thể là số âm

b) Các giá trị có khả năng xảy ra ở xác suất rất thấp. Ví dụ: 120 tuổi.


> Đối với dữ liệu hạng mục:

c) Cách cách nhập dữ liệu khác nhau, ví dụ như "tp Hồ Chí Minh" và "TPHCM"

d) Lỗi chính tả

e) Các giá trị có tần số xuất hiện rất thấp

f) Các giá trị không xuất hiện trong tập giá trị

Inter Quantile Range (IQR)

▶ Ngoài ra, ta có thể dùng IQR (khoảng tứ phân vị) để xác định phạm vi phần lớn dữ liệu tập trung `[lower,upper]` và xử lý các ngoại lệ nằm bên ngoài đoạn này bằng phương pháp **Quantile based flooring and capping**. Các giá trị nằm ngoài đoạn `[lower,upper]` sẽ được gán cho giá trị của upper nếu giá trị của nó > upper và lower nếu giá trị của nó < lower.

```python
from typing import Tuple
from sklearn.base import BaseEstimator, TransformerMixin

def find_boxplot_boundaries(
    col: pd.Series, whisker_coeff: float = 1.5
) -> Tuple[float, float]:
    """Findx minimum and maximum in boxplot.

    Args:
        col: a pandas serires of input.
        whisker_coeff: whisker coefficient in box plot
    """
    Q1 = col.quantile(0.25)
    Q3 = col.quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - whisker_coeff * IQR
    upper = Q3 + whisker_coeff * IQR
    return lower, upper


class BoxplotOutlierClipper(BaseEstimator, TransformerMixin):
    def __init__(self, whisker_coeff: float = 1.5):
        self.whisker = whisker_coeff
        self.lower = None
        self.upper = None

    def fit(self, X: pd.Series):
        self.lower, self.upper = find_boxplot_boundaries(X, self.whisker)
        return self

    def transform(self, X):
        return X.clip(self.lower, self.upper)
```

```python
clipped_absences = BoxplotOutlierClipper().fit_transform(df["absences"])

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15, 5))
clipped_absences.hist(bins=40, ax=axes[0])
clipped_absences.to_frame().boxplot(ax=axes[1], vert=False);
```

Meaning of covarience (hiệp phương sai) of 2 set X and Y - cov(X,Y):
- Là độ đo sự biến thiên cùng nhau của hai biến ngẫu nhiên (phân biệt với phương sai - đo mức độ biến thiên của một biến)
- Hiệp phương sai thể hiện mối tương quan giữa sự thay đổi của 2 biến. Nếu 2 biến có xu hướng thay đổi cùng nhau (nghĩa là, khi một biến có giá trị cao hơn giá trị kỳ vọng thì biến kia có xu hướng cũng cao hơn giá trị kỳ vọng), thì hiệp phương sai giữa hai biến này có giá trị dương.   image.png 

Meaning of Correlation Coefficient (hệ số tương quan):
- Thể hiện mối quan hệ giữa 2 biến là “mạnh” hay “yếu”.
- Hệ số tương quan có giá trị từ -1.0 đến 1.0.
- Hệ số tương quan có giá trị âm cho thấy hai biến có mối quan hệ nghịch biến hoặc tương quan âm.
- Hệ số tương quan có giá trị dương cho thấy mối quan hệ đồng biến hoặc tương quan dương.

⚓**Giới thiệu về kiểm định chi bình phương:**

- Dùng để kiểm tra xem có mối quan hệ giữa 2 biến (2 yếu tố) đang xét hay không. Tổng quát hơn, chi-square test kiểm tra liệu phân phối của categorical variables có khác so với biến còn lại. 
- Nhu cầu kiểm tra mối liên hệ giữa 2 biến xuất hiện nhiều trong các phương pháp phân tích như hồi quy Logistics, Survival analysis,... 
- Công thức: 

chi-square-formula.jpg

Với O là giá trị quan sát, E là giá trị mong đợi và c là bậc tự do.

** Cách tính giá trị mong đợi:
https://giamcanherbalthin.com/chi-square-la-gi/ 
- Nếu giá trị chi-square nhỏ, có mối tương quan cao giữa 2 tập dữ liệu.
- Nếu giá trị chi-square đủ lớn (>10), thì khả năng sẽ không có mối quan hệ nhiều giữa 2 tập dữ liệu.

** Xác định mối tương quan dựa vào giả thuyết H0 và p-value:  
- p-value: xác suất mắc phải sai lầm nếu ta bác bỏ giả thiết H0 (2 tiêu chuẩn phân loại độc lập với nhau). Nếu p-value càng bé, khả năng H0 được bác bỏ càng cao, tức là 2 biến có mối tương quan với nhau. Để xác định được p-value, ta cần có giá trị bậc tự do và chỉ số alpha.
- Giá trị p-value càng nhỏ (p-value < α) thì giả thuyết không càng nên bị bác bỏ và giả thuyết thay thế càng đáng tin cậy. 
- Giá trị p-value càng lớn (p-value> α) thì cơ sở để bác bỏ giả thuyết không càng yếu và kiểm định không có kết luận. 


```python
# chi-square test result
import statsmodels.api as sm
# convert contingency table into appropriate format to apply api of statsmodel library
romance_table = sm.stats.Table(romance_tab1)
# conduct chi-square test 
romance_rslt = romance_table.test_nominal_association()
romance_rslt.pvalue
```

✅ **Nhận xét**: Giá trị p-value khá nhỏ. Do đó chúng ta có thể kết luận rằng có mối liên hệ chặt chẽ giữa romantic status and final grade của các sinh viên.

**OLS regression model (Mô hình hồi quy OLS)** 

https://stataguide.wordpress.com/2020/04/13/mo-hinh-hoi-quy-ols/
- OLS - Ordinary Least Square (pp hồi quy bình phương tối thiểu) là phương pháp hồi quy được sử dụng phổ biến nhất trong nghiên cứu.
image.png

- Y và X được sử dụng để thể hiện cho tất cả quan sát của một tổng thể (population), còn y và x thể hiện cho các quan sát trong mẫu nghiên cứu được chọn (sample). Khi đó, các hệ số alpha và beta được ký hiệu dấu mũ - các giá trị ước lượng. 
- OLS sẽ lựa chọn các hệ số hồi quy alpha và beta sao cho tổng bình phương sai số của mô hình ước lượng là nhỏ nhất.
image.png
-  Có 3 vấn đề mà chúng ta cần quan tâm đầu tiên: đó là hệ số hồi quy có ý nghĩa thống kê không, mô hình có ý nghĩa không và mức độ giải thích của mô hình như thế nào?

- B1: kiểm định giả thuyết hệ số hồi quy. Mục tiêu là bác bỏ giả thuyết null (beta = 0). Để thực hiện kiểm định này, ta xem xét giá trị p-value tương ứng. Quan sát tóm tắt kết quả, ta thấy các biến father_education và mother_education đều có ảnh hưởng tích cực lên biến phụ thuộc final_score.
- B2: kiểm định mô hình hay còn gọi là kiểm định F (Giả thuyết cho kiểm định này là tất cả các hệ số hồi quy đồng thời bằng 0). Dùng thống kê F.
- B3: kiểm tra giá trị của R-square. R2 thể hiện cho % biến thiên của biến phụ thuộc được giải thích bởi mô hình. 

```python
# use GLS to see coefficients
edu = sm.GLS(y_edu, X_edu)
results_edu_wls = edu.fit()
results_edu_wls.summary()
```

Chúng ta sẽ sử dụng một loại biểu đồ mới đó là **Violin plot**

**Violin plot**:
- Biểu đồ violin tương tự như biểu đồ hộp, ngoại trừ việc chúng hiển thị mật độ xác suất của dữ liệu ở các giá trị khác nhau. Các biểu đồ này bao gồm một điểm đánh dấu cho trung vị của dữ liệu và một nến hiển thị phạm vi liên phần tư (giữa Q1 và Q3), như trong các boxplot.
- được sử dụng để thể hiện sự so sánh của một phân phối thay đổi (hoặc phân phối mẫu) trên các "danh mục" khác nhau. 

```python
# first, let's try observing the variation of studytime according to age 
plt.figure(figsize=(10,10))
sns.violinplot(x='age', y='study_time', data=df_main, palette='rocket')
plt.title('Study time By Age', fontsize=20)
plt.ylabel('Study time', fontsize=16)
plt.xlabel('Age', fontsize=16)
```

```python
# combine with design for higher education
plt.figure(figsize=(12,8))
sns.violinplot(x='age', y='study_time', hue='desire_higher_edu', data=df_main, palette="Accent_r")
plt.title('Distribution Of Study Time By Age & Desire To Receive Higher Education', fontsize=20)
plt.ylabel('Study Time', fontsize=16)
plt.xlabel('Age', fontsize=16)
```

Tiếp theo, nhóm sẽ sử dụng bar plot để phân tích phần trăm sinh viên có mong muốn được học lên cao hơn.


```python
higher_tab = pd.crosstab(index=df_main.final_grade, columns=df_main.desire_higher_edu)
higher_perc = higher_tab.apply(perc).reindex(index)
```

```python
higher_perc
```

```python
higher_perc.plot.bar(colormap="winter", figsize=(14,6), fontsize=16)
plt.title('Final Grade By Desire to Receive Higher Education', fontsize=20)
plt.xlabel('Final Grade', fontsize=16)
plt.ylabel('Percentage of Student', fontsize=16)
```

```python
study_tab = pd.crosstab(index=df.study_time, columns=df.final_grade)
study_perc = study_tab.apply(perc)
study_perc.plot.bar(colormap="winter", figsize=(14,6), fontsize=16)
plt.title('Final Grade By Study Time', fontsize=20)
plt.xlabel('Study Time', fontsize=16)
plt.ylabel('Percentage of Student', fontsize=16)
```

```python
# chi-square test result 
import statsmodels.api as sm
higher_table = sm.stats.Table(higher_tab)
higher_rslt = higher_table.test_nominal_association()
higher_rslt.pvalue
```

✅ **Nhận xét:** Mong muốn học đại học cũng như thời gian học có ảnh hưởng tương quan đến kết quả học tập của sinh viên.

#### 6) Final Grade By Living Area
> Kiểm tra sự khác nhau trong kết quả học tập các sinh viên sống ở thành phố và các sinh viên trong ở vùng ngoại ô


```python
# living area: urban vs. rural
df.address = df.address.map({'U':'Urban', 'R':'Rural'})
```

```python
plt.figure(figsize=(6,6))
sns.countplot(df.address, palette='summer')
plt.title('Urban and Rural Students Count', fontsize=20)
plt.xlabel('Living Area', fontsize=16)
plt.ylabel('Number Of Students', fontsize=16)
plt.show()
```

```python
ad_tab1 = pd.crosstab(index=df.final_grade, columns=df.address)
ad_tab = np.log(ad_tab1)
ad_perc = ad_tab.apply(perc).reindex(index)
```

```python
ad_perc.plot.bar(colormap="winter", fontsize=16, figsize=(8,6))
plt.title('Final Grade By Living Area', fontsize=20)
plt.ylabel('Percentage of Logarithm Student#', fontsize=16)
plt.xlabel('Final Grade', fontsize=16)
```

```python
# chi-square test result
ad_table = sm.stats.Table(ad_tab1)
ad_rslt = ad_table.test_nominal_association()
ad_rslt.pvalue
```

✅**Nhận xét:** Living area cũng có ảnh hưởng tương đối đến kết quả học tập của sinh viên.
