```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import KFold
```

```python
mush = pd.read_csv("E:\\Coding\\MachineLearning\\CSVFile\\mushrooms.csv", delimiter=';',skiprows=1)
plt.style.use('seaborn-bright')
mush.info()
```

```python
mush = mush.rename(columns={'class': 'myClass'})
mush = mush.rename(columns={'cap-shape': 'capshape'})
mush = mush.rename(columns={'cap-surface': 'capsurface'})
mush = mush.rename(columns={'cap-color': 'capcolor'})
mush = mush.rename(columns={'gill-attachment': 'gillatt'})
mush = mush.rename(columns={'gill-spacing': 'gillspa'})
mush = mush.rename(columns={'gill-size': 'gillsize'})
mush = mush.rename(columns={'gill-color': 'gillcolor'})
mush = mush.rename(columns={'stalk-shape': 'stalkshape'})
mush = mush.rename(columns={'stalk-root': 'stalkroot'})
mush = mush.rename(columns={'stalk-surface-above-ring': 'stalksuraboving'})
mush = mush.rename(columns={'stalk-surface-below-ring': 'stalksurbelowing'})
mush = mush.rename(columns={'stalk-color-above-ring': 'stalkcolaboving'})
mush = mush.rename(columns={'stalk-color-below-ring': 'stalkcolbelowing'})
mush = mush.rename(columns={'veil-type': 'veiltype'})
mush = mush.rename(columns={'veil-color': 'veilcolor'})
mush = mush.rename(columns={'ring-number': 'ringnum'})
mush = mush.rename(columns={'ring-type': 'ringtype'})
mush = mush.rename(columns={'spore-print-color': 'spocol'})
```

```python
mush.drop(mush[mush["stalkroot"] == '?'].index, inplace=True)
```

```python
#column = ['myClass', 'capshape','capsurface', 'capcolor', 'bruises','odor', 'gillatt', 'gillspa','gillsize', 'gillcolor', 'stalkshape','stalkroot', 'stalksuraboving','stalksurbelowing','stalkcolaboving' ,'stalkcolbelowing', 'veiltype','veilcolor','ringnum'	,'ringtype','spocol','population','habitat']
features = ['capshape','capsurface', 'capcolor', 'bruises','odor', 'gillatt', 'gillspa','gillsize', 'gillcolor', 'stalkshape','stalkroot', 'stalksuraboving','stalksurbelowing','stalkcolaboving' ,'stalkcolbelowing', 'veiltype','veilcolor','ringnum'	,'ringtype','spocol','population','habitat']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score
```

```python
models = []
models.append(('KNN', KNeighborsClassifier()))
models.append(('SVC', SVC()))
models.append(('LR', LogisticRegression(max_iter=5644)))
models.append(('DT', DecisionTreeClassifier()))
models.append(('GNB', GaussianNB()))
models.append(('RF', RandomForestClassifier()))
models.append(('GB', GradientBoostingClassifier()))
```

```python
mush.replace('a', 1, inplace=True)
mush.replace('b', 2, inplace=True)
mush.replace('c', 3, inplace=True)
mush.replace('d', 4, inplace=True)
mush.replace('e', 5, inplace=True)
mush.replace('f', 6, inplace=True)
mush.replace('g', 7, inplace=True)
mush.replace('h', 8, inplace=True)
mush.replace('i', 9, inplace=True)
mush.replace('j', 10, inplace=True)
mush.replace('k', 11, inplace=True)
mush.replace('l', 12, inplace=True)
mush.replace('m', 13, inplace=True)
mush.replace('n', 14, inplace=True)
mush.replace('o', 15, inplace=True)
mush.replace('p', 16, inplace=True)
mush.replace('q', 17, inplace=True)
mush.replace('r', 18, inplace=True)
mush.replace('s', 19, inplace=True)
mush.replace('t', 20, inplace=True)
mush.replace('u', 21, inplace=True)
mush.replace('v', 22, inplace=True)
mush.replace('w', 23, inplace=True)
mush.replace('x', 24, inplace=True)
mush.replace('y', 25, inplace=True)
mush.replace('z', 26, inplace=True)
```

```python
X = mush[features]
y = mush.myClass
```

```python
mush
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify = mush.myClass, random_state=0)
```

```python
X_train
```

```python
y_train
```

```python
names = []
scores = []
for name, model in models:
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    scores.append(accuracy_score(y_test, y_pred))
    names.append(name)
tr_split = pd.DataFrame({'Name': names, 'Score': scores})
print(tr_split)
```
